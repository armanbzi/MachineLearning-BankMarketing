```python
import pandas as pd
import requests

# Load names file
url_names = "https://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.names"
text = requests.get(url_names).text

# Extract feature lines (after @attribute section)
features = []
for line in text.split("\n"):
    if line.startswith("@attribute"):
        feature_name = line.split()[1]
        features.append(feature_name)

# Print all features
for i, f in enumerate(features):
    print(f"{i}: {f}")
```

```python
import pandas as pd
import requests
import zipfile
import io

# Correct UCI dataset URL (ZIP file)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip"

# Download and unzip in memory
response = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(response.content))

# List files inside zip (optional check)
print(z.namelist())

# Load the main dataset file
with z.open("bank-additional/bank-additional-full.csv") as f:
    df = pd.read_csv(f, sep=';')

# Print all feature names
print("\nFeatures:\n")
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")

print("\nShape:", df.shape)
```

```python
import pandas as pd

# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip"

import requests, zipfile, io

response = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(response.content))

with z.open("bank-additional/bank-additional-full.csv") as f:
    df = pd.read_csv(f, sep=';')

# Class distribution
class_counts = df['y'].value_counts()
class_percent = df['y'].value_counts(normalize=True) * 100

print("Class counts:\n", class_counts)
print("\nClass percentages:\n", class_percent)

# Imbalance ratio
ratio = class_counts['no'] / class_counts['yes']
print("\nImbalance ratio (no/yes):", ratio)
```
